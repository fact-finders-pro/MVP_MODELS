````python
def missing_report_for_group(df: pd.DataFrame, cols: list) -> pd.DataFrame:
    """
    Compute missing-value stats for a column group.

    Returns a DataFrame with:
      - column: column name
      - missing_count: number of missing values
      - missing_pct: percentage of missing values
    Sorted from highest to lowest missing_pct.
    """
    # Keep only columns that actually exist in df
    valid_cols = [c for c in cols if c in df.columns]

    total_rows = len(df)
    data = []
    for c in valid_cols:
        missing_count = df[c].isna().sum()
        missing_pct = (missing_count / total_rows * 100) if total_rows > 0 else 0.0
        data.append({
            "column": c,
            "missing_count": int(missing_count),
            "missing_pct": missing_pct
        })

    report = pd.DataFrame(data)
    return report.sort_values(by=["missing_pct", "missing_count"], ascending=False).reset_index(drop=True)
````

checking **multicollinearity** for `description` and `short_description`
# **Multi Collinearity**
Multicollinearity is a situation that happens in statistics when two or more
factors you’re using to make a prediction contain very similar information.

````python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from statsmodels.stats.outliers_influence import variance_inflation_factor
````

````python
def build_description_features(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    # Fill NaNs with empty strings for text processing
    df["description_filled"] = df["description"].fillna("")
    df["short_description_filled"] = df["short_description"].fillna("")

    # Basic numeric proxies
    df["description_len"] = df["description_filled"].str.len()
    df["short_description_len"] = df["short_description_filled"].str.len()
    df["description_word_count"] = df["description_filled"].str.split().str.len()
    df["short_description_word_count"] = df["short_description_filled"].str.split().str.len()

    return df
````

````python
def check_multicollinearity_basic(df_features: pd.DataFrame):
    cols = [
        "description_len", "short_description_len",
        "description_word_count", "short_description_word_count"
    ]
    X = df_features[cols].astype(float)

    # Correlation matrix
    corr = X.corr()
    print("Correlation matrix:\n", corr)

    # VIF for each feature
    vif_data = []
    for i in range(X.shape[1]):
        vif = variance_inflation_factor(X.values, i)
        vif_data.append({"feature": cols[i], "vif": vif})

    vif_df = pd.DataFrame(vif_data)
    print("\nVIF:\n", vif_df)
````

````python
df_desc = build_description_features(df)
check_multicollinearity_basic(df_desc)
````

````python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
````

````python
def cosine_similarity_desc_short(
    df: pd.DataFrame,
    desc_col: str = "description",
    sdesc_col: str = "short_description",
    min_df: int = 5,
    max_features: int = 20000
):
    # Fill NaNs
    desc_texts = df[desc_col].fillna("").astype(str)
    sdesc_texts = df[sdesc_col].fillna("").astype(str)

    # Build shared TF–IDF vocabulary on both columns
    corpus = pd.concat([desc_texts, sdesc_texts], axis=0)
    vectorizer = TfidfVectorizer(min_df=min_df, max_features=max_features)
    vectorizer.fit(corpus)

    X_desc = vectorizer.transform(desc_texts)
    X_sdesc = vectorizer.transform(sdesc_texts)

    # Row-wise cosine similarity between description and short_description
    sim_matrix = cosine_similarity(X_desc, X_sdesc)
    sim_row = sim_matrix.diagonal()

    # For rows where both texts are empty, set similarity to NaN
    both_empty = (desc_texts.str.len() == 0) & (sdesc_texts.str.len() == 0)
    sim_row = pd.Series(sim_row)
    sim_row[both_empty.values] = np.nan

    # Result DataFrame
    df_sim = pd.DataFrame(
        {
            desc_col: desc_texts,
            sdesc_col: sdesc_texts,
            "cosine_similarity": sim_row,
        }
    )

    # Aggregate stats
    summary = {
        "mean_similarity": sim_row.mean(),
        "median_similarity": sim_row.median(),
        "min_similarity": sim_row.min(),
        "max_similarity": sim_row.max(),
        "non_null_pairs": int(sim_row.notna().sum()),
    }

    return df_sim, summary
````

First, the cosine similarity results show that, on average, `description` and `short_description` are moderately similar (mean and median around 0.44–0.45), with many pairs being very close (similarity near 1). This tells me they generally talk about the same core idea of the startup, so using `short_description` when `description` is missing is not introducing random or unrelated information. At the same time, they are not perfect duplicates, which is good, because for rows where both exist, the model can still pick up additional nuance from the full description.

Second, from a modelling and data-quality perspective, leaving `description` as null for 12% of rows would waste a valuable text signal, especially when I have a clean, always-present alternative in `short_description`.

I get the best of both worlds: rich long descriptions where available, and a meaningful fallback where they are missing, plus an explicit flag telling the model that the original long description was not provided.
